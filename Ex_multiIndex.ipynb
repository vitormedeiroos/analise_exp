```python
import pandas as pd
```

```python
arrays = [[1,2],['azul', 'verde']]
```

#MultiIndex em Arrays

```python
pd.MultiIndex.from_arrays(arrays, names=('number', 'color'))
```

#MultiIndex em produto cartesiano

```python
numbers = [0, 1, 2]
colors = ['green', 'purple']
pd.MultiIndex.from_product([numbers, colors],
                           names=['number', 'color'])
```
