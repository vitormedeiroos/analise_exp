#Pivot Table

A funçao pivot nn consegue manipular indices com valores repetidos

````python
carros = [4,3,4,5,5]
dias = pd.date_range('20190101','20190101', periods=5)
vendedor = ['Vitor', 'João', 'Pedro', 'Nicolas', 'Vitor',]

df = pd.DataFrame({ 'Vendas': carros, 'Data': dias, 'Vendedor':vendedor})
````

````python
df
````

````python
pd.pivot(df, index='Data',columns='Vendedor', values='Vendas')
````

````python
pd.pivot_table(df, index='Data',columns='Vendedor', values='Vendas')#(padrão) - faz a média dos valores repetidos
#caso queira mudar... add o atributo, aggfunc = 'sum'
````

````python
pd.pivot_table(df, index='Data',columns='Vendedor', values='Vendas', aggfunc='sum')
````
